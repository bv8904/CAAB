```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
import shutil

def prepare_data(source_dir, base_dir):
    """
    Labels UTKFace images as 'child' or 'adult' and organizes them into folders.
    """
    data = []
    for filename in os.listdir(source_dir):
        if filename.endswith(".jpg"):
            try:
                # UTKFace format: [age]_[gender]_[race]_[date].jpg
                age = int(filename.split('_')[0])
                label = 'child' if age < 18 else 'adult'
                data.append({'filename': filename, 'label': label})
            except (IndexError, ValueError):
                continue

    df = pd.DataFrame(data)
    train_df, val_df = train_test_split(df, test_size=0.2, stratify=df['label'], random_state=42)

    # Create directory structure
    for split, split_df in [('train', train_df), ('val', val_df)]:
        for label in ['child', 'adult']:
            os.makedirs(os.path.join(base_dir, split, label), exist_ok=True)
        
        # Copy files to organized directories
        for _, row in split_df.iterrows():
            src = os.path.join(source_dir, row['filename'])
            dst = os.path.join(base_dir, split, row['label'], row['filename'])
            shutil.copy(src, dst)

    print(f"Dataset organized successfully at: {base_dir}")

# Antigravity Prompt: "Agent, run this script to organize /content/utkface into /content/data"
# prepare_data('/content/utkface', '/content/data')
```
